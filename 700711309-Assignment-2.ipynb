```python
#question1
#1.1 Using NumPy create random vector of size 15 having only Integers in the range 1-20
arr = np.random.randint(1,20,15) #creating random vector with integers in the range 1 to 20
arr   #diplay array
```

```python
# 1.2 Reshape the array to 3 by 5
arr.resize((3,5)) #reshaping above array using resize method
arr    #diplay array
```

```python
# 1.3 Print array shape
arr.shape # shape function is used to print the array shape
```

```python
# 1.4 Replace the max in each row by 0
for i in arr:      #using for-loop for locating max in each row
    m = i.argmax()
    i[m] = 0
arr   #diplay array
```

```python
#question 2
# 2.1 Read the provided CSV file ‘data.csv’
df = pd.read_csv('data.csv') #reading the file using pandas
# print(df)
print(df.to_string()) #print the csv file and convert to string


```

```python
#2.2 Show the basic statistical description about the data.
df.describe() #using describe() function to show statistical description about data 
```

```python
# 2.3 Check if the data has null values.
df.isnull().any() #Using isnull() function to check if any null values are present
```

```python
# 2.3.1 Replace the null values with the mean
df.fillna(df.mean(),axis=0,inplace=True) #replacing the null values with mean value using fillna() function
print(df.to_string()) # printing the data
```

```python
#2.4 Select at least two columns and aggregate the data using: min, max, count, mean.
df[['Duration','Pulse']].aggregate(func=['min','max','count','mean'],axis=0,)    #using aggregate() function to calculate 
                                                                                 #for 2 columns
```

```python
# 2.5 Filter the dataframe to select the rows with calories values between 500 and 1000. 
df[(df['Calories']>=500) & (df['Calories']<=1000)] #Using data frames within data frames 
                                                   #and filtering the values between 500 & 1000
```

```python
# 2.6 Filter the dataframe to select the rows with calories values > 500 and pulse < 100. 
df[(df['Calories']>500)&(df['Pulse']<100)] # Using data frames within data frames and filtering the values > 500 & < 100
```

```python
# 2.7 Create a new “df_modified” dataframe that contains all the columns from df except for “Maxpulse”.
df_modified = df[['Duration','Pulse','Calories']] #Creating a new dtaframe without Maxpulse col with name df_modified.
df_modified #print df_modified
```

```python
# 2.8 Delete the “Maxpulse” column from the main df dataframe
df.drop(labels = ['Maxpulse'],axis=1,inplace=True) #deleting Maxpulse col using drop() function
df   #diplay df
```

```python
# 2.9 Convert the datatype of Calories column to int datatype.
df['Calories'] = df['Calories'].astype(int)   #using astype() function for converting datatype into int
df.dtypes    #printing datatypes
```

```python
# 2.10 Using pandas create a scatter plot for the two columns (Duration and Calories).
df.plot.scatter(x='Duration',y='Calories')    #using plot.scatter() 
```

```python
#question 3
#Sample data:
#Programming languages: Java, Python, PHP, JavaScript, C#, C++
#Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
#3.1
%matplotlib inline
from matplotlib import pyplot as plt    #importing pyplot
programming_languages = ['Java','Python','PHP','JavaScript','C#','C++']     #creating a array of prog languages
popularity = [22.2,17.6,8.8,8,7.7,6.7]    #creating array for popularity
explode = [0.1, 0, 0, 0, 0, 0]    #giving explode - 0.1 for Java only based on sample data
plt.pie(popularity, labels=programming_languages, explode=explode, shadow=True,startangle=135,
        autopct='%1.1f%%',      #using autopct for displaying percentage values
        wedgeprops={'edgecolor': 'black'})    #creating pie diagram with necessary parameters
plt.show()
```
